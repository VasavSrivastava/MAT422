#**4.2 Graph Modeling**

Graph modeling involves representing a system or network as a graph $G(V, E)$, where $V = \{v_1, v_2, \dots, v_n\}$ is the set of nodes, and $E = \{e_1, e_2, \dots, e_m\}$ is the set of edges. Edges may be directed or undirected and can have weights representing the strength of connections. Graphs can be represented by an adjacency matrix $A$, where $A_{ij} = 1$ if $v_i$ and $v_j$ are connected, and $0$ otherwise. The Laplacian matrix $L$ is defined as $L = D - A$, where $D$ is the degree matrix with $D_{ii} = \sum_{k} E_{ik}$. Eigenvalues and eigenvectors of $L$ provide insights into graph properties. For example, the smallest eigenvalue $\lambda_1 = 0$ corresponds to $Le = 0$, where $e = [1, 1, \dots, 1]^T$. The second smallest eigenvalue $\lambda_2$ (Fiedler value) and its corresponding eigenvector $x$ minimize $x^T L x$, subject to $x^T x = 1$ and $x^T e = 0$. The adjacency matrix for directed graphs is generally asymmetric, whereas for undirected graphs, it is symmetric $(A = A^T)$. Finally, the incidence matrix $I_G$ and its transpose can decompose $L$ as $L = I_G \cdot I_G^T$.


```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from scipy.linalg import eigh

# Create a simple graph
G = nx.Graph()
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1), (2, 4)])

# Plot the graph
plt.figure(figsize=(6, 6))
nx.draw(G, with_labels=True, node_color='lightblue', node_size=800)
plt.title("Graph Visualization")
plt.show()

# Get the adjacency matrix
A = nx.adjacency_matrix(G).toarray()
print("Adjacency Matrix:\n", A)

# Compute the Laplacian matrix
L = nx.laplacian_matrix(G).toarray()
print("Laplacian Matrix:\n", L)

# Compute eigenvalues and eigenvectors of the Laplacian matrix
eigenvalues, eigenvectors = eigh(L)
print("Eigenvalues:\n", eigenvalues)
print("Eigenvectors (columns):\n", eigenvectors)

# Verify Fiedler vector (eigenvector corresponding to second smallest eigenvalue)
fiedler_vector = eigenvectors[:, 1]
print("Fiedler Vector:\n", fiedler_vector)
```
