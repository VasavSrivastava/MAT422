#####**2.2.1.1 Sample spaces and events**
A sample space is the set of all possible outcomes of a random experiment, representing the universe of everything that could happen in that scenario. An event is a subset of the sample space, consisting of one or more outcomes that share a common characteristic or interest. For example, if the sample space represents the possible outcomes of rolling a die, the event "rolling an even number" includes the outcomes {2, 4, 6}. The probability of an event is the measure of how likely that event is to occur within the sample space.

```python
import random

# Define the sample space for a six-sided die
sample_space = [1, 2, 3, 4, 5, 6]

# Define an event: rolling an even number
event_even = [2, 4, 6]

# Simulate rolling the die
roll = random.choice(sample_space)

# Check if the roll is part of the event
if roll in event_even:
    print(f"Rolled {roll}: It's an even number (Event occurred).")
else:
    print(f"Rolled {roll}: It's an odd number (Event did not occur).")
```

#**2.2.2 Conditional Probability**
Conditional probability is the probability of an event occurring given that another event has already occurred. It quantifies how the likelihood of an event changes when additional information is known. Mathematically, if \(A\) and \(B\) are two events, the conditional probability of \(A\) given \(B\) is denoted as \(P(A|B)\) and is calculated using the formula $$
P(A|B) = \frac{P(A \cap B)}{P(B)}, \quad \text{where } P(B) > 0
$$
Below is an example that considers a situation where we have a deck of cards, and we want to calculate the probability of drawing a King given that the card drawn is a face card (Jack, Queen, or King).

```python
import random

# Define the deck of cards (simplified with just face cards)
deck = ['Jack', 'Queen', 'King'] * 4  # 4 suits, 3 face cards each

# Total number of face cards
total_face_cards = len(deck)

# Number of Kings in the deck
total_kings = deck.count('King')

# Define the event: drawing a face card
face_card = random.choice(deck)

# Conditional probability calculation
if face_card in ['Jack', 'Queen', 'King']:
    if face_card == 'King':
        conditional_probability = 1
    else:
        conditional_probability = 0
    print(f"Drawn card: {face_card}")
    print(f"Conditional Probability of King given that a face card is drawn: {total_kings / total_face_cards}")
else:
    print(f"Drawn card: {face_card}")
    print("The card drawn is not a face card, conditional probability is not applicable.")
```

#**2.2.3 Discrete Random Variables**

A discrete random variable is a type of random variable that can take on a countable number of distinct values, typically integers, each associated with a certain probability. Unlike continuous random variables, which can take any value within a range, discrete random variables have specific, separated outcomes, such as the result of rolling a die or the number of heads in a series of coin flips. The probabilities of all possible values of a discrete random variable sum up to 1, and its probability distribution is often represented by a probability mass function (PMF). Below is an example that demonstrates the concept of a discrete random variable using the scenario of rolling a six-sided die.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the possible outcomes of rolling a six-sided die
possible_outcomes = [1, 2, 3, 4, 5, 6]

# Define the probability distribution for a fair die (each outcome has an equal probability)
probabilities = [1/6] * 6

# Simulate rolling the die 1000 times
num_rolls = 1000
rolls = np.random.choice(possible_outcomes, size=num_rolls, p=probabilities)

# Count the frequency of each outcome
outcome_counts = {outcome: list(rolls).count(outcome) for outcome in possible_outcomes}

# Display the results
print("Outcome counts after rolling the die 1000 times:")
for outcome, count in outcome_counts.items():
    print(f"Outcome {outcome}: {count} times")

# Plot the probability mass function (PMF)
plt.bar(outcome_counts.keys(), outcome_counts.values(), color='blue')
plt.xlabel('Die Face')
plt.ylabel('Frequency')
plt.title('Discrete Random Variable: Rolling a Six-Sided Die')
plt.show()
```

#####**2.2.3.1 The expected value and variance of X**
The expected value of a discrete random variable \(X\), denoted as \(E(X)\), is a measure of the central tendency or the average value that \(X\) is expected to take over many trials of the experiment. It is calculated by summing the products of each possible value of \(X\) and its corresponding probability, using the formula: $$ E(X) = \sum_{i} x_i \cdot P(X=x_i)
$$ where \(x_i\) are the possible values of \(X\) and \(P(X=x_i)\) is the probability of \(X\) taking the value \(x_i\). The variance of \(X\), denoted as \(\text{Var}(X)\), measures the spread or dispersion of the values of \(X\) around the expected value. It is calculated using the formula: $$ \text{Var}(X) = E[(X-E(X))^2] = \sum_{i} (x_i - E(X))^2 \cdot P(X=x_i) $$ which reflects how much the values of \(X\) deviate from the mean on average.


```python
import numpy as np

# Define the possible outcomes of rolling a six-sided die
possible_outcomes = np.array([1, 2, 3, 4, 5, 6])

# Define the probability distribution for a fair die (each outcome has an equal probability)
probabilities = np.array([1/6] * 6)

# Calculate the expected value E(X)
expected_value = np.sum(possible_outcomes * probabilities)

# Calculate the variance Var(X)
variance = np.sum(((possible_outcomes - expected_value) ** 2) * probabilities)

# Display the results
print(f"Expected Value (E(X)): {expected_value}")
print(f"Variance (Var(X)): {variance}")
```

#**2.2.4 Continuous Random Variables**
A continuous random variable is a type of random variable that can take on an infinite number of possible values within a given range or interval. Unlike discrete random variables, which have distinct, separate values, continuous random variables can assume any value within a continuum, such as any real number between two bounds. The probability of a continuous random variable taking any specific value is zero; instead, probabilities are determined over intervals using a probability density function (PDF).

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define the parameters of the normal distribution
mean = 0      # Mean of the distribution
std_dev = 1   # Standard deviation of the distribution

# Generate 1000 random values from the normal distribution
random_values = np.random.normal(loc=mean, scale=std_dev, size=1000)

# Plot the distribution of the random values
sns.histplot(random_values, kde=True)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Continuous Random Variable: Normal Distribution')
plt.show()
```
